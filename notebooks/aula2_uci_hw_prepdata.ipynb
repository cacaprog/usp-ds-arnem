```python
import pandas as pd
import os
```

```python
features_path = os.path.join('/home/usp-ds-arnem/data/uci', 'features.txt')
```

```python
features = pd.read_csv(features_path, 
                       sep='\s+',
                       header=None, names=['index', 'feature'])
```

```python
features
```

```python
# remover caracteres especiais (isso pode dar uma dor de cabeça depois...)
features = features.feature.str.replace('(','')\
        .str.replace(')','')\
        .str.replace(',','')\
            .to_list()
```

```python
# Verificando os nomes das variáveis
nomes_variaveis = [f'{i:03}-{features[i]}' for i in range(len(features))]
```

```python
nomes_variaveis
```

```python
pd.Series(features).duplicated().sum()
```

Carregar arquivos 'train'

```python
# Definir o caminho dos arquivos de forma portátil
X_train_path = os.path.join('/home/usp-ds-arnem/data/uci', 'train', 'X_train.txt')
subject_train_path = os.path.join('/home/usp-ds-arnem/data/uci', 'train', 'subject_train.txt')
```

```python
# Ler os arquivos
X_train = pd.read_csv(X_train_path, sep='\s+', header=None)
subject_train = pd.read_csv(subject_train_path, sep='\s+', header=None)

# Definir os nomes das colunas
subject_train.columns = ['subject']
X_train.columns = nomes_variaveis

X_train = pd.concat([subject_train, X_train], axis=1)

X_train.head()
```

```python
# Definir o caminho dos arquivos de forma portátil
activity_labels_path = os.path.join('/home/usp-ds-arnem/data/uci', 'activity_labels.txt')
y_train_path = os.path.join('/home/usp-ds-arnem/data/uci', 'train', 'y_train.txt')

# Ler o arquivo features.txt
activity_labels = pd.read_csv(activity_labels_path, 
                              sep='\s+', 
                              header=None, 
                              index_col = 0,
                              names=['label'])

# Ler o arquivo X_train.txt, especificando o delimitador como espaço
y_train = pd.read_csv(y_train_path, sep='\s+', header=None)

# Definir os nomes das colunas usando os nomes das variáveis do features.txt
activity_labels = activity_labels.to_dict()['label']

y_train.columns = ['label']
y_train.label.dtypes
activity_labels

# TRansforma y_train em categorical, com os rótulos corretos
y_train['label'] = pd.Categorical(y_train.label.map(activity_labels), ordered=False)

y_train.value_counts(dropna=False)
```

```python
# Ler as bases de testes
X_test_path = os.path.join('/home/usp-ds-arnem/data/uci', 'test', 'X_test.txt')
subject_test_path = os.path.join('/home/usp-ds-arnem/data/uci', 'test', 'subject_test.txt')


# Ler o arquivo X_test.txt, especificando o delimitador como espaço
X_test = pd.read_csv(X_test_path, sep='\s+', header=None)

# Ler o arquivo subject_test.txt, especificando o delimitador como espaço
subject_test = pd.read_csv(subject_test_path, sep='\s+', header=None)
subject_test.columns = ['subject']

# Definir os nomes das colunas usando os nomes das variáveis do features.txt
X_test.columns = nomes_variaveis

# juntar X_train com subject_train
X_test = pd.concat([subject_test, X_test], axis=1)#.set_index(['subject'], append=True)

# Exibir as primeiras linhas do dataframe
X_test.head()
```

```python
# Ler y_train
# Definir o caminho dos arquivos de forma portátil
y_test_path = os.path.join('/home/usp-ds-arnem/data/uci', 'test', 'y_test.txt')

# Ler o arquivo X_test.txt, especificando o delimitador como espaço
y_test = pd.read_csv(y_test_path, sep='\s+', header=None)

y_test.columns = ['label']

# TRansforma y_train em categorical, com os rótulos corretos
y_test['label'] = pd.Categorical(y_test.label.map(activity_labels), ordered=False)

y_test.value_counts(dropna=False)
```
